```python
import math
time = 1001
x = np.array([0])
y = np.array([0])

for i in range(time):
    x = np.append(x, i)
    
for i in range(time):
    tim = 0.01* math.exp(.05*i)
    y = np.append(y,tim)
```

```python
plt.plot(x,y)
plt.xlabel("Years")
plt.ylabel("quintillions")
plt.title("countiniously compounded")
plt.show
```

```python
x2 = np.array([0])
y2 = np.array([0])

for i in range(time):
    x2 = np.append(x2, i)
    
for i in range(time):
    mit = 0.01* (1+(0.05*i))
    y2 = np.append(y2,mit)
```

```python
print(x2)
```

```python
print(y2)
```

```python
plt.plot(x2,y2)
plt.xlabel("Years")
plt.ylabel("money")
plt.title("Simple Interest")
plt.show
```

```python
plt.plot(x,y-y2)
plt.xlabel("Years")
plt.ylabel("quintillions")
plt.title("Difference between simple and compoundedcontinuosuly interest")
plt.show
```

```python
import random 
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

def random_walkloop(numberofloops):
    i = 0
    
    for k in range(numberofloops):
        init_position = (0,0)
        num_steps = 10000

        x = np.array([init_position[0]]*num_steps)
        y = np.array([init_position[1]]*num_steps)
        options = ['left','right','down','up']

        for i in range(num_steps):

            direction = random.choice(options)
                #print(direction)

            if direction == 'left':            
                    x[i] = x[i - 1]- 1
                    y[i] = y[i - 1]
            elif direction == 'right':
                    x[i] = x[i - 1]+ 1
                    y[i] = y[i - 1]
            elif direction == 'down':
                    x[i] = x[i - 1]
                    y[i] = y[i - 1]- 1
            elif direction == 'up':
                    x[i] = x[i - 1]
                    y[i] = y[i - 1]+ 1
            else:
                    print("Wrong")
            #print(x)    
            #print(y)

        plt.plot(x,y)
        plt.show()
```

```python
random_walkloop(5)
```

What do you notice?
There seems to be a large area covered by the random walk, and it does not seem to be concentrated in one area. 

```python
def random_walkloopweights(numberofloops):
    i = 0
    
    for k in range(numberofloops):
        init_position = (0,0)
        num_steps = 10000

        x = np.array([init_position[0]]*num_steps)
        y = np.array([init_position[1]]*num_steps)
        options = ['left','right','down','up']

        for i in range(num_steps):

            direction1 = random.choices(['left','right','down','up'], weights =(.1,.1,.1,.7))
            #print(type(direction1))
            direction = direction1[0]
           # direction = str(direction)
            #print(type(direction))
            
            if direction == 'left':
                #add y or else you get triangle
                    x[i] = x[i - 1]- 1
                    y[i] = y[i - 1]
            elif direction == 'right':
                    x[i] = x[i - 1]+ 1
                    y[i] = y[i - 1]
            elif direction == 'down':
                    x[i] = x[i - 1]
                    y[i] = y[i - 1]- 1
            elif direction == 'up':
                #add x or else you get triangle
                    x[i] = x[i - 1]
                    y[i] = y[i - 1]+ 1
            else:
                    print("Wrong")
            #print(x)    
            #print(y)

        plt.plot(x,y)
        plt.show()
```

```python
random_walkloopweights(5)
```

What do you notice?
I see that all of the graphs indicate that the person is moving upward, this is proabably due to the weights. Compared to the other unweighted graphs, this one seems to be moving only upward, and does not cover a large area.
