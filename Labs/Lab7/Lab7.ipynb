```python
import random
def randomwalk(num_steps,start_position):
    
    options = [-1,1]
    directions = []
    i = 0
    
    while i < num_steps:
        
        direction = random.choice(options)
       # print(direction)
        directions.append(direction)
        
        i += 1
        
    final = start_position + sum(directions)
    return final
    
def simulaterandomwalk(num_sims,num_steps,start_position):
    i = 0
    lst = []
    
    while i < num_sims:
        walk = randomwalk(num_steps,start_position)
        lst.append(walk)
        i += 1
        
    return np.array(lst)   

```

```python
simulaterandomwalk(3,5,0)
```
