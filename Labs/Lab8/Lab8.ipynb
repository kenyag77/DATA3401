```python
from sklearn.decomposition import PCA
import matplotlib
from matplotlib import pyplot as plt

plt.figure()
PCA3=PCA(n_components=2)
XPCA = PCA3.fit_transform(x)
plt.plot(XPCA[y==0,0],XPCA[y==0,1],'g.')
plt.plot(XPCA[y==1,0],XPCA[y==1,1],'r.')
plt.show()
```
